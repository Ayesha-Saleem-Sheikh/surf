```python
# imported python libraries
import pickle
import pandas as pd
```

```python
# opened the pkl file and loaded it into data
with open('grein_count_matrix.pkl', 'rb') as file:
    data = pickle.load(file)
```

```python
# printed the data 
print(data)
```

```python
# named all the columns
# list(data.columns)
```

```python
# converted the data into panda dataframe
# printed the dataframe
df = pd.DataFrame(data)
print(df)
```

```python
# saving as tsv file
df.to_csv('grein_count_matrix.tsv', sep="\t")
```

```python
# checked if the number of unique values in the gene column is equal to the index number
len(df['gene'].unique()) == len(df.index)
```

```python
# checked if the number of unique values in gene_symbol column is equal to the index number
len(df['gene_symbol'].unique()) == len(df.index)
```
