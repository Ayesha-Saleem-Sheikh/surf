# CODE BY CODE!!

```python
# function that calculates the idf 
import math

def calculate_idf(corpus,words):
    idf = {}
    D = len(corpus)

    for word in set(words):
        df_t = 0
        for each_title in corpus:
            words_in_title = each_title.lower().split()
            if word in words_in_title:
                df_t +=1
        if df_t>0:
            idf[word] = math.log((D)/(df_t))

    return (idf)
```

```python
#Calculating the tf
def calculate_tf(title):
    tf = {}
    x = title.lower().split()
    t_words = len (x)
    
    for word in x :
        tf [word] = tf.get(word,0) +1

    for word in tf:
        tf [word] = tf [word] / t_words
    return tf 
```

```python
# Calculate tf_idf 

def calculate_tf_idf(tf,idf):
    tfidf = {}
    for word in tf :
        tfidf[word] = tf [word] * idf.get(word,0)
    return (tfidf)
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

def lemm(text):
    lemmtexts = []
    for  title in text:
        doc = nlp(title)
        lemmas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
        lemmtext = ' '.join(lemmas)
        lemmtexts.append(lemmtext)


    return lemmtexts
    
            
```
